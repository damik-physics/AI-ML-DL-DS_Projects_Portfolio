```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# show all cell outputs in notebook
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all" 

# Read data into a pandas DataFrame
df = pd.read_csv('../data/lung_cancer.csv')


# refactor column names removing spaces and special characters
df.columns = df.columns.str.replace(' ', '_')
# capitalize column names
df.columns = df.columns.str.capitalize()

# Get the first 5 rows of the DataFrame
df.head()

# Get the last 5 rows of the DataFrame
df.tail()

# Get the number of rows and columns
df.shape

# Get the column names
df.columns

# Get the data types of the columns
df.dtypes
```

```python

# Get the number of unique values in each column
df.nunique()

# Get the number of missing values in each column
df.isnull().sum()

# Get the number of non-missing values in each column
df.count()


# Get a summary of the statistical properties of the data
df.describe()

# Get information about the DataFrame
df.info()
```

```python

# Get the frequency counts of the Gender column
df['Gender'].value_counts()

# Visualize the data
df.plot(kind='bar', x='Gender', y='Age', title='Age by Gender', figsize=(10, 10))
plt.show()

# Compute pairwise correlations
corr = df.corr()

# Visualize the correlations using a heatmap
plt.figure(figsize=(10, 10))
plt.imshow(corr, cmap='RdYlGn', interpolation='none', aspect='auto')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr)), corr.columns)
plt.suptitle('Correlations Heatmap', fontsize=15, fontweight='bold')
plt.show()
```

* Use the groupby() and mean() methods to compute the mean values of numerical variables for different groups, such as the different age groups or smoking status.

* Use the pivot_table() method to create a cross-tabulation of two categorical variables, such as Gender and Smoking.

* Use the scatter() method to visualize the relationship between two numerical variables, such as Age and Chronic Disease.

* Use the boxplot() method to visualize the distribution of a numerical variable across different groups, such as Age by Smoking.

* Use the seaborn library to create more sophisticated plots, such as violin plots, swarm plots, and pair plots.

```python

# Compute the mean values of Age by Smoking status
df.groupby('Smoking')['Age'].mean()

# Create a cross-tabulation of Gender and Smoking
pd.pivot_table(df, index='Gender', columns='Smoking',
               values='Age', aggfunc='mean')

```

```python

# Visualize the relationship between Age and Chronic Disease
df.plot(kind='scatter', x='Age', y='Chronic_disease')
plt.show()

# Visualize the distribution of Age by Smoking status using a boxplot
sns.boxplot(x='Smoking', y='Age', data=df)
plt.show()

# Visualize the distribution of Age using a violin plot
sns.violinplot(x='Smoking', y='Age', data=df)
plt.show()

# Visualize the relationship between all pairs of variables using a pair plot
sns.pairplot(df, hue='Smoking')
plt.show()
```
